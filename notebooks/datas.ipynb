Salvando Dados Do Orçamento Municipal - 2010 à 2024

```python
import requests
import pandas as pd
import os

url_base = "https://api-dados-abertos.tce.ce.gov.br/dados_orcamentos?codigo_municipio=048&exercicio_orcamento={ano}00"

anos = list(range(2010, 2025))

pasta_base = 'dados_orcamentos'

os.makedirs(pasta_base, exist_ok=True)


for ano in anos:

    url = url_base.format(ano=ano)
    try:
        res = requests.get(url)

        dados = res.json()

        if "data" not in dados or not dados["data"]:
            print(f"Nenhum dado encontrado para ano {ano}")
            continue

        df = pd.DataFrame(dados["data"])

        caminho_arquivo = os.path.join(pasta_base, f"dados_{ano}.csv")
        df.to_csv(caminho_arquivo, index=False)

        print(f"Dados do ano {ano} salvos com sucesso!")

    except Exception as e:
        print(f"Erro ao processar o ano {ano}: {e}")




```
